```python
# heatmap: to calc pair wise similarity
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sample data for demonstration (replace this with your actual data)
documents = [
    "This is a sample document.",
    "This document is another example.",
    "We are testing similarity using TF-IDF.",
    "Similarity matrices are useful in NLP tasks.",
    "Another document for testing."
]

# Create a TF-IDF matrix
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# Convert TF-IDF matrix to a DataFrame
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out()
)

# Calculate pairwise cosine similarity
similarity_matrix = cosine_similarity(tfidf_df)

# Convert the similarity matrix to a DataFrame for easier plotting
similarity_df = pd.DataFrame(similarity_matrix)

# Create a heatmap of the similarity matrix
plt.figure(figsize=(10, 8))
sns.heatmap(
    similarity_df,
    cmap="coolwarm",  # Diverging colormap for better contrast
    annot=True,       # Display similarity values in cells
    fmt=".2f",        # Format values to 2 decimal places
    linewidths=0.5,   # Add lines between cells
    cbar_kws={"label": "Similarity Score"}  # Label for the color bar
)

# Customize titles and labels
plt.title("Pairwise Similarity Heatmap", fontsize=16)
plt.xlabel("Document Index", fontsize=12)
plt.ylabel("Document Index", fontsize=12)

# Adjust tick labels for better readability
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10, rotation=0)

# Show the plot
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```

```python
import pandas as pd
import tensorflow_hub as hub
import numpy as np

# Load the dataset
file_path = '/content/dataset-tickets-multi-lang3-4k.csv'
df = pd.read_csv(file_path)

# Preprocess the "answer" column
df['answer'] = df['answer'].fillna("").astype(str)  # Handle missing values and ensure text format

# Load the Universal Sentence Encoder (USE) model
use_model = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

# Function to generate embeddings
def embed_answers(answers):
    embeddings = use_model(answers)  # Generate embeddings
    return embeddings.numpy()  # Convert TensorFlow tensors to numpy arrays

# Embed the 'answer' column in batches to handle memory efficiently
batch_size = 100
embeddings = []

for i in range(0, len(df), batch_size):
    batch = df['answer'][i:i + batch_size].tolist()
    batch_embeddings = embed_answers(batch)
    embeddings.append(batch_embeddings)

# Concatenate all batches into a single numpy array
embeddings = np.vstack(embeddings)

# Add embeddings to the DataFrame as a new column
df['answer_embeddings'] = list(embeddings)

# Save the updated dataset to a new CSV file
output_path = '/content/dataset-tickets-multi-lang3-4k.csv'
df.to_csv(output_path, index=False)

print(f"Updated dataset with embeddings saved to: {output_path}")
```
